```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.yelp.com/biz/tian-tian-hainanese-chicken-rice-singapore-7?osq=Restaurants'
html = requests.get(url)
s = BeautifulSoup(html.content, 'html.parser')

ratings = s.find_all(class_="css-10n911v")
for rating in ratings:
  print(rating.text)
```

```python
dateList = []

for i in range(0, 200, 10):
  url = 'https://www.yelp.com/biz/tian-tian-hainanese-chicken-rice-singapore-7?osq=Restaurants' + '&start=' + str(i)
  html = requests.get(url)
  s = BeautifulSoup(html.content, 'html.parser')

  dates = s.find_all(class_="css-10n911v")
  for date in dates:
    dateList.append(date.text)

for item in dateList:
  print(item)
```

```python
updated_dateList = []
for i in range(0, 200):
  current = dateList[i][:12]
  if current[-1].isdigit() == False:
    current = current[:len(current)-1]
  updated_dateList.append(current)

for date in updated_dateList:
  print(date)
```

```python
years = []
for date in updated_dateList:
  years.append(date[-4:])

for year in years:
  print(year)
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('assignment2_tiantian_reviews.csv')

# Display the DataFrame
df.head()
```

```python
df['date'] = updated_dateList
df['year'] = years

df.head()
```
